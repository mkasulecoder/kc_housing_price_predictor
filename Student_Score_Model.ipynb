```python
"""
Letâ€™s build a linear regression model that predicts the scores based on its number of hours for study features.
@author Mark Kasule
Introduction to Data Science [DATA 300]
Date 07/01/2024
"""
# import all libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# LR2 DATASET
# Read file from current working directory
file = "student_scores.csv"
df = pd.read_csv(file)
print(df.head(10))
```

```python
"""
Lets declare the X (features) = Hours, and y (Predicted) = Scores
"""
X = df["Hours"].values.reshape(-1, 1) # Reshape X to be 2D
y = df["Scores"]
```

```python
"""
Graphing the original raw data
"""
sns.scatterplot(x=X.flatten(), y=y, color="red") # convert X back to 1-dimensional
plt.legend(["Scores"],loc="best")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.title("Scores Earned from Number of Hours Spent Reading")
sns.set(rc={'axes.facecolor':'lightblue', 'figure.facecolor':'lightgreen'})
plt.show()
```

```python
"""
Let's make score supervised predictions with sklearn based existing data provided
"""
from sklearn.model_selection import train_test_split

# Split the data from the train and test
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0, train_size=0.8)
```

```python
"""
Let's import Linear Regression Model, Fit the model, and make predictions
"""
from sklearn.linear_model import LinearRegression

# Fit the model
lr = LinearRegression()
lr.fit(X_train, y_train)

# Make predictions
y_predictions = lr.predict(X_test)
```

```python
"""
Plot the Actual scores vs the predicted scores for the test set
DataFrame is always 2D, while a Series is 1D.
"""
# Access X_test data not as a Dataframe `X_test["Hours"]` since its 2D now
# To access its elements, we use slicing instead with flatten()
plt.scatter(X_test.flatten(), y_test, color="blue", label="Actual Scores")
plt.scatter(X_test.flatten(), y_predictions, color="red", label="Predicted Scores")
plt.ticklabel_format(axis="y", style="plain")
plt.ylabel("Scores")
plt.xlabel("Hours")
plt.title("Scores Earned from Number of Hours Spent Reading")
#plt.legend()
plt.show()
```
